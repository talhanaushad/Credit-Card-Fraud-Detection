```python
data = data[["Car_Owner","Propert_Owner", "Annual_income", "EDUCATION","Family_Members","label"]]
```

```python
data
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
data["Car_Owner"]= labelencoder.fit_transform(data["Car_Owner"])
data["Propert_Owner"]= labelencoder.fit_transform(data["Propert_Owner"])
data["EDUCATION"]= labelencoder.fit_transform(data["EDUCATION"])
```

```python
data
```

```python
data.drop(columns="Family_Members", inplace=True)
```

```python
data
```

```python
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
def modelperformance(predictions):
    print(f"The accuracy score on the model is {accuracy_score(Y_test,predictions)}")
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(scaled_X_train,Y_train)
```

```python
log_performance = log_model.predict(scaled_X_test)

```

```python
modelperformance(log_performance)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors" : [3,5,7,9,11],
               "weights" : ["uniform","distance"],
               "algorithm" :["auto","kd_tree"]
               }
```

```python
gridkn = GridSearchCV(KNeighborsClassifier(),param_grid,cv=2)
```

```python
gridkn.fit(scaled_X_test,Y_test)
```

```python
gridkn.best_params_
```

```python
knprids = gridkn.predict(scaled_X_test)
```

```python
modelperformance(knprids)
```

```python
from sklearn.svm import SVC
svm = SVC()
param_grid_svr = {"C": [0.01,0.1,0.5,1],
                  "kernel": ["linear","rbf","poly"]
                   }
grid_svr = GridSearchCV(svm,param_grid_svr)
```

```python
grid_svr.fit(scaled_X_train,Y_train)
```

```python
pred_svc = grid_svr.predict(scaled_X_test)
modelperformance(pred_svc)
```

```python
from sklearn.tree import DecisionTreeClassifier
param_grid = {"criterion":["gini","entropy"],
             "splitter" : ["best","random"],
             "max_depth": [None,10,20,30,40,50],
             "min_samples_split":[2,5,10],
             "min_samples_leaf": [1,2,4]
}
```

```python
grid_search_tree = GridSearchCV(estimator=DecisionTreeClassifier(),param_grid=param_grid)
```

```python
grid_search_tree.fit(scaled_X_train,Y_train)
```

```python
grid_search_tree.best_params_
```

```python
preds = grid_search_tree.predict(scaled_X_test)
```

```python
modelperformance(preds)
```
